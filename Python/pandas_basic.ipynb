### 1. Numpy를 활용한 csv file open

```python
def file_open_by_numpy():
    # np.loadtxt(구분자 = ',', 데이터 타입: string)
    np_arr = np.loadtxt('data/test_data.CSV', delimiter=",", encoding='cp949', dtype=str)
    return np_arr
```

```python
arr = file_open_by_numpy()
print(arr)
```

#### 3.7. 특정 데이터 조회
- `iloc()` 함수 사용
  - `row 번호`로 조회 가능
- `loc()` 함수 사용
  - `index명`으로 조회 가능

```python
df.loc[2]  # 특정 행 조회
df.loc[[2, 3]]  # 여러 행 조회
df.loc[1:4]  # 특정 구간 조회
df.loc[df.index != 2]  # 특정 행 제외하고 조회
df.loc[:, '이름']  # 특정 열 조회
df.loc[:, df.columns != '나이']  # 특정 열 제외하고 조회
df.loc[:, '나이':'직업']  # 특정 범위의 열 조회
df.loc[[2, 5], ['이름', '직업']]  # 행과 열 조회
```

#### 3.8. 행 삽입하기
#### 3.8.1. 마지막 위치에 하나의 행 추가
- `loc()` 사용
- 하나의 리스트를 마지막에 추가

```python
df.loc[len(df)] = ['마지막', '99', '남', '무직', '서울/경기']
df.tail(5)
```

#### 3.8.2. 두 개의 DataFrame 합치기
- `concat()` 사용
- 딕셔너리 형태를 DataFrame 에 추가하고 싶을 때 사용
- `ignore_index=True` 로 설정해야 인덱스가 기존 행의 뒷 번호로 잘 지정됨

```python
df2 = pd.DataFrame({
    '이름': ['진짜마', '지막임'],
    '나이': ['98', '97'],
    '성별': ['여', '남'],
    '직업': ['개발자', '모델'],
    '사는곳': ['전라북도', '충청남도'] 
})
df2
```

```python
new_df = pd.concat([df, df2], ignore_index=True)
new_df.tail(5)
```

#### 3.8.4. 중간에 데이터 삽입
- `concat()` & `iloc` 사용
- `pd.concat([df.iloc[:N], 원하는 행, df.iloc[N:]], ignore_index=True)`
  - N 번째에 새로운 DataFrame을 넣고 싶을 때

```python
new_df = pd.concat([df.iloc[:3], df2, df.iloc[3:]], ignore_index=True)
new_df.head(10)
```

#### 3.12. DataFrame 복사
- 대입 연산자 사용 시 얕은 복사
- 원본 데이터 변경을 원치 않을 시 `copy()` 함수를 통해 **깊은 복사**

```python
# 데이터 새로 가져오기
arr = file_open_by_numpy()
columns=arr[0]
arr = np.delete(arr, 0, 0)
df = pd.DataFrame(arr, columns=columns)
```

```python
# 얕은 복사
df2 = df
df2['나이'] = 10
df.head()
```

```python
# 깊은 복사
df2 = df.copy()
df2['나이'] = 10
df.head()
```
